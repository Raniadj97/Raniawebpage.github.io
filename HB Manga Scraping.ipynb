```python
from bs4 import BeautifulSoup
import requests
import time
import datetime

import smtplib #you can send emails to urself with this librairie
```

```python
URL= 'https://hbmangakissa.com/products/dragon-ball-la-terrible-muscle-tower-t05'
headers = {"User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/120.0.0.0 Safari/537.36"}
page = requests.get(URL , headers = headers)
```

```python
soup1 = BeautifulSoup (page.content , 'html.parser')
soup2 = BeautifulSoup(soup1.prettify(), 'html.parser') #to make the output looks better
print(soup2)
```

```python
#title = soup2.find('div', attrs = {'class': 'product__title' }).get_text()
title = soup2.find("a", class_ = 'product__title').get_text()
price = soup2.find('span', attrs= {'class' : 'price-item price-item--regular'}).get_text()
print(title)
print(price)
```

```python
price = price.strip()
title = title.strip()
print(price)
print(title)
```

```python
price = price.strip()[3:]
print(price)
```

```python
#create a timestamp
date = datetime.date.today()
print(date)
```

```python
#create a csv file
import csv

header = ['title', 'price', 'date']
data = [title, price, date]

with open ('hbmangadataset.csv', 'w', newline='', encoding= 'UTF8') as f:
    writer = csv.writer(f)
    writer.writerow(header)
    writer.writerow(data)

```
